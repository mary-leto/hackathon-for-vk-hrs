```python
!pip install python-docx PyPDF2 pdfminer.six striprtf

import re
import sys
import os
import json
from typing import Dict, List, Tuple

# Импорт tkinter, не завершаем работу, если его нет
try:
    import tkinter as tk
    from tkinter import filedialog
    _TKINTER_AVAILABLE = True
except ImportError:
    _TKINTER_AVAILABLE = False

# -------------------------------
# Словарь категорий навыков (можно пополнять и корректировать)
# -------------------------------
SKILL_CATEGORIES = {
    "Programming Languages": [
        "Python", "Java", "C++", "C#", "JavaScript", "TypeScript", "PHP", "Go",
        "Ruby", "Swift", "Kotlin", "Rust", "Scala", "Perl"
    ],
    "Web Frameworks": [
        "HTML", "CSS", "JavaScript", "React", "Angular", "Vue.js", "Django", "Flask",
        "Spring", ".NET Core", "Laravel", "Express", "Node.js", "Next.js", "Nuxt.js", "Svelte"
    ],
    "Databases": [
        "SQL", "MySQL", "PostgreSQL", "SQLite", "Oracle",
        "MongoDB", "Redis", "Cassandra", "Firebase", "MariaDB"
    ],
    "Version Control & DevOps": [
        "Git", "GitHub", "GitLab", "SVN",
        "Docker", "Kubernetes", "Jenkins", "CI/CD",
        "Bash", "Linux", "GitHub Actions", "Azure DevOps"
    ],
    "Cloud Platforms": [
        "AWS", "Azure", "Google Cloud", "Heroku", "DigitalOcean", "Yandex.Cloud"
    ],
    "Data & ML": [
        "Pandas", "NumPy", "TensorFlow", "PyTorch", "Scikit-learn", "Jupyter",
        "Keras", "OpenCV", "NLTK", "SpaCy"
    ],
    "IT Practices": [
        "OOP", "REST API", "Microservices", "Agile", "Scrum", "Kanban", "TDD",
        "Unit Testing", "BDD", "CI/CD"
    ],
    "Soft Skills": [
        # — Communication —
        "Communication", "Коммуникация", "Общение", "Communicative", "Communicative skills",
        "Коммуникативный", "Коммуникабельный",
        # — Teamwork —
        "Teamwork", "Team Work", "Работа в команде", "Командная работа",
        "Team-oriented", "Team oriented", "Командный", "Командный игрок",
        # — Problem Solving —
        "Problem Solving", "Решение проблем", "Навыки решения проблем",
        "Problem-solving", "Проблемно-ориентированный",
        # — Critical Thinking —
        "Critical Thinking", "Критическое мышление", "Critical-thinking", "Критически мыслящий",
        # — Adaptability —
        "Adaptability", "Адаптивность", "Гибкость", "Adaptable", "Адаптивный", "Гибкий",
        # — Self-Learning —
        "Self-Learning", "Самообучение", "Способность к самообучению", "Self-learning", "Самообучающийся",
        # — Time Management —
        "Time Management", "Тайм-менеджмент", "Управление временем", "Time-management",
        "Организованный", "Умеющий планировать время",
        # — Creativity —
        "Creativity", "Креативность", "Creative", "Креативный",
        # — Collaboration —
        "Collaboration", "Сотрудничество", "Коллаборация", "Collaborative",
        "Сотрудничающий", "Коллаборативный",
        # — Emotional Intelligence —
        "Emotional Intelligence", "Эмоциональный интеллект", "EQ",
        "Emotionally intelligent", "Эмоционально интеллигентный", "Эмоционально интеллектуальный",
        # — Resilience —
        "Resilience", "Устойчивость", "Стрессоустойчивость", "Resilient",
        "Устойчивый", "Стойкий",
        # — Growth Mindset —
        "Growth Mindset", "Установка на рост", "Ориентированный на рост",
        "Growth-minded", "Рост-ориентированный",
        # — Remote Work —
        "Remote Work", "Удалённая работа", "Дистанционная работа", "Remote-working",
        "Удалённо работающий", "Работа на удалёнке",
        # — Digital Literacy —
        "Digital Literacy", "Цифровая грамотность", "Digitally literate",
        "Грамотный в цифровых технологиях", "Цифрово грамотный"
    ]
}


def preprocess_text(text: str) -> str:
    """
    1. Переводим в нижний регистр.
    2. Убираем все символы, кроме букв (русских и латиницы), цифр, пробелов и дефиса.
    3. Сжимаем множественные пробелы до одного.
    """
    text = text.lower()
    text = re.sub(r'[^a-zа-я0-9\s\-]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text


def check_categories(
    resume_text: str,
    skill_categories: Dict[str, List[str]]
) -> Tuple[Dict[str, bool], Dict[str, List[str]]]:
    """
    Для каждой категории проверяет, встречается ли хотя бы одно ключевое слово из списка.
    Возвращает:
      - category_matched: {название_категории: True/False}
      - found_keywords:   {название_категории: [список найденных слов]}
    """
    preprocessed = preprocess_text(resume_text)
    category_matched = {}
    found_keywords = {}

    for category, keywords in skill_categories.items():
        found_in_cat = []
        for kw in keywords:
            kw_norm = preprocess_text(kw)
            if kw_norm and kw_norm in preprocessed:
                found_in_cat.append(kw)
        category_matched[category] = len(found_in_cat) > 0
        found_keywords[category] = found_in_cat

    return category_matched, found_keywords


def analyze_resume(resume_text: str) -> None:
    """
    Основная функция:
      1. Проверяет каждую категорию навыков по триггер-словам.
      2. Считает количество найденных ключевых слов и долю от общего количества.
      3. Выводит вердикт: соответствует ли резюме условным требованиям или нужно доработать.
      4. Выводит статистику по категориям и общую процентную «полезность» резюме.
    """
    category_matched, found_keywords = check_categories(resume_text, SKILL_CATEGORIES)

    # Подсчитываем общее количество ключевых слов и найденных
    total_keywords = sum(len(words) for words in SKILL_CATEGORIES.values())
    total_found = sum(len(found) for found in found_keywords.values())
    overall_percentage = (total_found / total_keywords) * 100 if total_keywords else 0.0

    # Подсчёт по категориям: сколько найдено и сколько всего
    category_stats = {}
    for category, keywords in SKILL_CATEGORIES.items():
        found_count = len(found_keywords[category])
        total_count = len(keywords)
        pct = (found_count / total_count) * 100 if total_count else 0.0
        category_stats[category] = {
            "found_count": found_count,
            "total_count": total_count,
            "percentage": pct,
            "matched": category_matched[category],
            "found_list": found_keywords[category],
            "missing_list": [kw for kw in keywords if kw not in found_keywords[category]]
        }

    # Определяем вердикт: если хотя бы в одной категории ничего не найдено — рекомендуем доработку
    missing_categories = [cat for cat, matched in category_matched.items() if not matched]
    if not missing_categories:
        verdict = "Резюме соответствует всем требованиям — найдены ключевые навыки во всех категориях."
    else:
        verdict = (
            "Резюме требует доработки. В следующих категориях не найдено ни одного ключевого слова: "
            + "; ".join(missing_categories)
        )

    # Вывод результатов
    print("=== ВЕРДИКТ ===")
    print(verdict)
    print(f"\nОбщая «проверка по словам»: {total_found} / {total_keywords} "
          f"({overall_percentage:.1f}%)\n")

    print("=== СТАТИСТИКА ПО КАТЕГОРИЯМ ===")
    for cat, stats in category_stats.items():
        status = "✔" if stats["matched"] else "✘"
        print(f"\n{status} {cat}: {stats['found_count']} / {stats['total_count']} "
              f"({stats['percentage']:.1f}%)")
        print(f"    Найдено: {stats['found_list']}")
        missing = stats["missing_list"]
        if missing:
            preview = missing[:5]
            more = len(missing) - len(preview)
            ellipsis = f", …(+{more} ещё)" if more > 0 else ""
            print(f"    Отсутствует: {preview}{ellipsis}")
        else:
            print("    Отсутствует: нет\n")


def extract_text_from_docx(path: str) -> str:
    """
    Читает .docx и возвращает весь текст как одну строку.
    Если python-docx не установлен, выводит ошибку и завершает работу.
    """
    try:
        from docx import Document
    except ImportError:
        print("Ошибка: для обработки .docx необходимо установить python-docx (pip install python-docx).")
        sys.exit(1)

    doc = Document(path)
    paragraphs = [p.text for p in doc.paragraphs if p.text.strip()]
    return "\n".join(paragraphs)


def extract_text_from_pdf(path: str) -> str:
    """
    Читает .pdf и возвращает весь текст как одну строку.
    Если PyPDF2 не установлен, выводит ошибку и завершает работу.
    """
    try:
        import PyPDF2
    except ImportError:
        print("Ошибка: для обработки .pdf необходимо установить PyPDF2 (pip install PyPDF2).")
        sys.exit(1)

    text_parts = []
    with open(path, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            page_text = page.extract_text() or ""
            text_parts.append(page_text)
    return "\n".join(text_parts)


def extract_text_from_json(path: str) -> str:
    """
    Читает .json и возвращает объединенные все строковые значения.
    """
    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
    except Exception as e:
        print(f"Ошибка при чтении JSON: {e}")
        sys.exit(1)

    def recurse(obj):
        texts = []
        if isinstance(obj, dict):
            for v in obj.values():
                texts.append(recurse(v))
        elif isinstance(obj, list):
            for item in obj:
                texts.append(recurse(item))
        elif isinstance(obj, str):
            texts.append(obj)
        # остальные типы (int, float, None) игнорируем
        return " ".join(texts)

    return recurse(data)


def select_resume_file() -> str:
    """
    Пытаемся открыть графическое окно выбора файла (File Open Dialog).
    Если tkinter недоступен, нет дисплея, или DISPLAY не задан, возвращаем ввод пользователем пути вручную.
    """
    # Если tkinter не установлен или нет переменной DISPLAY
    if not _TKINTER_AVAILABLE or "DISPLAY" not in os.environ:
        return input("Введите путь к файлу резюме (.txt/.pdf/.docx/.json): ").strip()

    # Пробуем создать окно; в случае ошибок тоже запрашиваем путь вручную
    try:
        root = tk.Tk()
        root.withdraw()  # Скрываем главное окно
        filetypes = [
            ("Text files", "*.txt"),
            ("Word Documents", "*.docx"),
            ("PDF files", "*.pdf"),
            ("JSON files", "*.json"),
            ("All files", "*.*")
        ]
        path = filedialog.askopenfilename(
            title="Выберите файл с резюме",
            filetypes=filetypes
        )
        root.destroy()
        return path or ""
    except tk.TclError:
        return input("Введите путь к файлу резюме (.txt/.pdf/.docx/.json): ").strip()


if __name__ == "__main__":
    # Запрашиваем путь через диалог/консоль
    resume_path = select_resume_file()
    if not resume_path:
        print("Файл не выбран. Завершаем работу.")
        sys.exit(1)

    if not os.path.isfile(resume_path):
        print(f"Ошибка: файл '{resume_path}' не найден.")
        sys.exit(1)

    ext = os.path.splitext(resume_path)[1].lower()
    if ext == ".txt":
        try:
            with open(resume_path, "r", encoding="utf-8") as f:
                resume_text = f.read()
        except Exception as e:
            print(f"Ошибка при чтении .txt файла: {e}")
            sys.exit(1)

    elif ext == ".docx":
        resume_text = extract_text_from_docx(resume_path)

    elif ext == ".pdf":
        resume_text = extract_text_from_pdf(resume_path)

    elif ext == ".json":
        resume_text = extract_text_from_json(resume_path)

    else:
        print("Ошибка: неподдерживаемый формат. Поддерживаются: .txt, .pdf, .docx, .json")
        sys.exit(1)

    analyze_resume(resume_text)
```
